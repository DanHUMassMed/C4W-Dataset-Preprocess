```python
import pandas as pd

# Load the Excel file
file_path = "../data/external/us_zipcodes/ZIP_Locale_Detail.xls"
df = pd.read_excel(file_path)

# Filter for Worcester, MA
worcester_zips = df[
    (df["PHYSICAL CITY"].str.upper() == "WORCESTER")
    & (df["PHYSICAL STATE"].str.upper() == "MA")
]

# Extract ZIP codes
worcester_zip_codes = worcester_zips["PHYSICAL ZIP"].tolist()

# Display the ZIP codes
print(worcester_zip_codes)
```
